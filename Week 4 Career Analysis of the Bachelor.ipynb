```python
import numpy as np
import pandas as pd
import matplotlib as plt
```

```python
women = pd.read_csv("bachelor-contestants.csv")
men = pd.read_csv("bachelorette-contestants.csv")
```

```python
women = pd.DataFrame(women["Occupation"].value_counts())
display(women)
```

```python
men = pd.DataFrame(men["Occupation"].value_counts())
display(men)
```

```python
women = women.reset_index()
women.columns = ['Occupation', 'Occurances']
display(women)
```

```python
womenShort = women[:20]
display(womenShort)
```

```python
men = men.reset_index()
men.columns = ['Occupation', 'Occurances']
display(men)
```

```python
menShort = men[:20]
display(menShort)
```

```python
menShort = menShort.assign(Salary = [143200, 123030, 54650, 36830, 75180, 28000, 152610, 92330, 183060, 79520, 67780, 141890, 66100, 44740, 79000, 105000, 80360, 39140, 31710, 93090])
display(menShort)
```
